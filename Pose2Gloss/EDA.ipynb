```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./gloss_info.csv')
df.head()
```

```python
df[df.g_type=='WORD'].gloss.value_counts()
```

```python
g_list = []
for g in list(df[df.g_type=='SEN'].gloss):
    if g is not np.nan:
        g_list+=g.split('|')
```

```python
len(np.unique(g_list+list(df[df.g_type=='WORD'].gloss)))
```

```python
word2id = {'<pad>':0,
           '<sos>':1,
           '<eos>':2}
for i, word in enumerate(np.unique(g_list+list(df[df.g_type=='WORD'].gloss))):
    word2id[word] = i+3

id2word = {v:k for k, v in word2id.items()}
```

```python
word2id
```

```python
import json

with open('word2id.json', 'w') as json_file:
    json.dump(word2id, json_file, ensure_ascii=False)
with open('id2word.json', 'w') as json_file:
    json.dump(id2word, json_file, ensure_ascii=False)
```

```python
df[df.g_type=='SEN'].gloss
```

```python
gloss_valid = pd.read_csv('gloss_label_valid.csv')
```

```python
print(len(word2id), len(id2word))
for g in gloss_valid.gloss.values:
    for gg in g.split('|')[1:-1]:
        try:
            word2id[gg]
        except:
            i = len(word2id)
            word2id[gg] = i
            id2word[i] = gg
print(len(word2id), len(id2word))
```

```python
with open('./word2id.json', "r") as st_json: word2id = json.load(st_json)
with open('./id2word.json', "r") as st_json: id2word = json.load(st_json)
```
